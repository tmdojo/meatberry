```python
logfile = "20170718101507_2.7_0.005_0.001.csv"
time_t, temperature, pwr, tempe_bottom, P, I, D = np.loadtxt(logfile, delimiter=',', unpack=True)

fig = plt.figure(facecolor='white')
ax = fig.add_subplot('211')
ax.plot(time_t/60, temperature, label='temperature')
#ax.set_xlabel('Elapsed time (min)')
ax.set_ylabel('Temperature (degrees C)')
ax.set_title('Pot')
ax.grid('on')
ax.legend(loc='upper left', bbox_to_anchor=(1.0, 1))

ax2 = fig.add_subplot('212')
ax2.plot(time_t/60, pwr, label='power')
ax2.plot(time_t/60, P, label='P')
ax2.plot(time_t/60, I, label='I')
ax2.plot(time_t/60, D, label='D')
ax2.set_xlabel('Elapsed time (min)')
ax2.set_ylabel('Control')
#ax2.set_title('Pot')
ax2.grid('on')
ax2.legend(loc='upper left', bbox_to_anchor=(1.0, 1))

plt.tight_layout()
```
